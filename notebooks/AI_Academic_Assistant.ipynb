```python
from langchain_community.document_loaders import PyPDFLoader

documents = []

pdf_files = ["COA_notes.pdf", "OOP_notes.pdf"]

for pdf in pdf_files:
    loader = PyPDFLoader(pdf)
    documents.extend(loader.load())

print("Pages Loaded:", len(documents))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=3000,
    chunk_overlap=300
)

chunks = text_splitter.split_documents(documents)

print("Total Chunks:", len(chunks))
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-2"
)

print("Embeddings Loaded Successfully")
```

```python
test_chunks = chunks[:100]

print("Using", len(test_chunks), "chunks")
```

```python
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_documents(
    test_chunks,
    embeddings
)

print("FAISS Database Created Successfully")
```

```python
retriever = vectorstore.as_retriever(
    search_kwargs={"k": 3}
)

print("Retriever Created Successfully")
```

```python
query = "What is inheritance?"

docs = retriever.invoke(query)

print("Retrieved Chunks:", len(docs))

for i, doc in enumerate(docs):
    print(f"\n----- Chunk {i+1} -----\n")
    print(doc.page_content[:500])
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.3
)

print("Gemini Loaded Successfully")
```

```python
def ask_notes(question):

    docs = retriever.invoke(question)

    context = docs[0].page_content

    prompt = f"""
    Use only the note below.

    Note:
    {context}

    Question:
    {question}

    Answer:
    """

    response = llm.invoke(prompt)

    return response.content
```

```python
answer = ask_notes(
    "What is an assembler?"
)

print(answer)
```

```python
def summarize_notes(topic):

    docs = retriever.invoke(topic)

    context = docs[0].page_content

    prompt = f"""
    Summarize the following notes in simple language.

    Notes:
    {context}

    Summary:
    """

    response = llm.invoke(prompt)

    return response.content
```

```python
print(
    summarize_notes("Assembler")
)
```

```python
def generate_quiz(topic):

    docs = retriever.invoke(topic)

    context = docs[0].page_content

    prompt = f"""
    Generate 5 MCQs from the notes.

    Notes:
    {context}

    MCQs:
    """

    response = llm.invoke(prompt)

    return response.content
```

```python
print(
    generate_quiz("Assembler")
)
```

```python
while True:

    print("\nAI Academic Assistant")
    print("1. Ask Question")
    print("2. Summarize Topic")
    print("3. Generate Quiz")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        q = input("Question: ")
        print(ask_notes(q))

    elif choice == "2":
        t = input("Topic: ")
        print(summarize_notes(t))

    elif choice == "3":
        t = input("Topic: ")
        print(generate_quiz(t))

    elif choice == "4":
        break
```
